# Thomas BMT Equation

# $\frac{d\vec{s}}{dt} = \frac{q}{m \gamma} \vec{S} \times \big( (1+a\gamma)\vec{B}-a(\gamma - 1)(\hat{t}\cdot \vec{B})\hat{t}-(a\gamma + \frac{\gamma}{1+\gamma})\frac{\beta \hat{t}\times \vec{E}}{c}\big)$

# Where a is the anomalous g factor $a=\frac{g-2}{2};\hat{t}$ is the dirction aligned with the particle motion

# First, we are only aiming at the ring dipoles, so the second and the third term vanish

# Now the equation reduced to:  $\frac{d\vec{s}}{dt}=\vec{\Omega} \times \vec{S} = (\frac{1}{\gamma}+a)\frac{q}{m} \vec{S} \times\vec{B_{\bot}} =-(\frac{1}{\gamma}+a)\frac{e }{m} B_{\bot} \vec{S}\times \hat{y}=(\frac{1}{\gamma}+a)\frac{e }{m} B_{\bot}(S_z \hat{x}-S_x \hat{z})$

#  $\vec{\Omega} = -\frac{q}{m \gamma}(1+a\gamma)\vec{B}_{\bot}$,  $\vec{\Omega}$ is the axial vector which describes the spin precession(direction and the speed of precession)


# For electron, q = -e, where e is the elementary charge,  $\vec{\Omega} = \frac{e}{m \gamma}(1+a\gamma)\vec{B}_{\bot}$
# Note: For the negative charge particle like electron, the axial vector is always aligned with the external mangetic field (counterclockwise rotation following the right hand rule) ; For positive charge particle, it's always anti-aligned (clockwise rotation )



$ $
$ $

#  Let $\omega=\frac{e}{m \gamma}(1+a\gamma)B_{\bot} $
#  The first order equations of spin motion is given by :

# $\begin{cases}\begin{align*} \dot{S_x} &= \omega S_z \\ \dot{S_y} &=0 \\ \dot{S_z}&=-\omega S_x\end{align*} \end{cases}$          

# Take the derivative to get the second order equations of spin motion:

# $\begin{cases} \begin{align*} \ddot{S_x} &= \omega \dot{S}_z = -\omega^2 S_x \\\ddot{S_y} &=0 \\ \ddot{S_z} &= \omega \dot{S}_x = -\omega^2 S_z \end{align*}\end{cases}$

## simple harmonic oscillator in x and z,  solution is give by:

# $\begin{cases} 
        \begin{align*}
        S_x &= A\sin(\omega t + \phi) \\  
        S_y &= B \\
        S_z&= A\cos(\omega t+ \phi) 
        \end{align*}
     \end{cases}$

## give the initial condition $(\tilde{S}_x,\tilde{S}_y,\tilde{S}_z)$

# $\begin{cases} 
        \begin{align*}
            A &= \tilde{S}_x \sin\phi+\tilde{S}_{z} \cos\phi\\  
            B &= \tilde{S}_y  \\
            \phi &= \arctan\big(\frac {\tilde{S}_{x}}{\tilde{S}_{z}}\big)
        \end{align*}
     \end{cases}$



$  $


# When electron passes through the dipole field, the amount of spin rotation should be given by:   $\alpha=\omega\cdot \frac{L}{v}= (1+a\gamma)\frac{eL}{\gamma mv} B_{\bot}=(1+a\gamma)\frac{eL}{p} B_{\bot}$

# when $\gamma =1$ (particle at rest), it reduces to Larmor precession

# Where L is the length of the dipole; $\alpha$ is the amount of rotation; $E_0$ is the rest mass energy of the electron, unit is eV, which can cancel e in the numerator, all the rest quantities are in the SI unit.

$ $
$ $

# Now, consider the solenoid field, which is parallel to the  beam motion



# $\frac{d\vec{S}}{dt} = \frac{d\vec{S}}{ds} \frac{ds}{dt}= v \frac{d\vec{S}}{ds}\Rightarrow  \frac{d\vec{S}}{ds}= \frac{1}{v}\frac{ds}{dt}$

# where s is the longitudinal orbit in the co-moving frame

# $\begin{align} \frac{d\vec{s}}{ds} &= \frac{q}{\gamma m v} \vec{S} \times \big( (1+a\gamma)\vec{B}-a(\gamma - 1)(\hat{t}\cdot \vec{B})\hat{t} \big) \\ &= \frac{q}{p} \vec{S} \times \big( (1+a\gamma)(\vec{B}_{\bot}+\vec{B}_{//})-a(\gamma - 1)\vec{B}_{//} \big) \\ &= \frac{q}{p} \vec{S} \times \big( (1+a\gamma)\vec{B}_{\bot}+(1+a)\vec{B}_{//} \big)\\ \end{align}  $


$ $
$ $
$ $


# Let $\omega_{\bot}= \frac{e}{p}(1+a\gamma)B_{\bot}, \omega_{//} = \frac{e}{p}(1+a)B_{//}$, and q =-e for the electron 


# $\frac{d\vec{s}}{ds} = \big(\omega_{\bot} \hat{y} + \omega_{//} \hat{z}\big)\times \vec{S}    =  \big(-\omega_{//}S_{y} + \omega_{\bot}S_{z} \big)\hat{x}+\omega_{//}S_{x} \hat{y} - \omega_{\bot}S_{x} \hat{z}$ 


# we can conclude 3 differential equations of the spin motion:

# $\begin{cases} \begin{align*} \dot{S_x} &= \omega_{\bot} S_z-\omega_{//}S_y \\ \dot{S_y} &= \omega_{//} S_x \\ \dot{S_z}&=-\omega_{\bot} S_x\end{align*}\end{cases}$

# Take the derivative and get the second order equations:

# Let $\omega= \sqrt{{\omega_{//}}^2+{\omega_{\bot}}^2}$

# $\begin{cases} \begin{align*} \ddot{S_x} &= -\omega_{//}\dot{S}_y + \omega_{\bot} \dot{S}_z = -\omega^2 S_x \\ \ddot{S_y} &= \omega_{//} \dot{S}_x = -{\omega_{//}}^2 S_y + \omega_{//} \omega_{\bot} S_z \\ \ddot{S_z}&=-\omega_{\bot} \dot{S}_x= \omega_{//} \omega_{\bot} S_y-{\omega_{\bot}}^2 S_z \end{align*}\end{cases}$

# It's easy to solve the equation of motion for $S_x$


# $S_x = A\sin\big(\omega t + \phi \big)$


# y and z are coupled, use matrix to solve it

# $\big(\begin{smallmatrix}
  \ddot{S_y}\\
  \ddot{S_z}
\end{smallmatrix}\big)=\bigg(\begin{smallmatrix}
  -{\omega_{//}}^2 &\omega_{//}\omega_{\bot}\\
  \omega_{//}\omega_{\bot}& -{\omega_{\bot}}^2
\end{smallmatrix}\bigg)\big(\begin{smallmatrix}
  S_y\\
  S_z
\end{smallmatrix}\big) $

# Corresponding eigenvalues: $-{\omega_{//}}^2-{\omega_{\bot}}^2$ or 0

# The eigenvector of $-{\omega_{//}}^2-{\omega_{\bot}}^2$: $\frac{1}{\omega}\big(\begin{smallmatrix}
  -\omega_{//}\\
  \omega_{\bot}
\end{smallmatrix}\big)$, for 0, it is: $\frac{1}{\omega}\big(\begin{smallmatrix}
  \omega_{\bot}\\
  \omega_{//}
\end{smallmatrix}\big)$

# The solution for y-z motion can be written as :

# $\begin{cases} 
      S_y= -\frac{\omega_{//}}{\omega} C\cos(\omega s+\phi )+\frac{\omega_{\bot}}{\omega}D \\
      S_z= \frac{\omega_{\bot}}{\omega} C\cos(\omega s+\phi )+\frac{\omega_{//}}{\omega}D
   \end{cases}$

# Use the constrain that the length of spin vector is invariant, ${S_x}^2+{S_y}^2+{S_z}^2=1$ , can get A=C

# The general solution is given by:
# $\begin{cases} 
        S_x = C \sin{(\omega s + \phi)} \\
        S_y= -\frac{\omega_{//}}{\omega} C\cos(\omega s+\phi )+\frac{\omega_{\bot}}{\omega}D\\
        S_z= \frac{\omega_{\bot}}{\omega} C\cos(\omega s+\phi )+\frac{\omega_{//}}{\omega}D
   \end{cases}$

$ $
$ $


# Given initial condition spin vector is vertical (0,1,0), can get:

# $\begin{cases} 
        C = -\frac{\omega_{//}}{\omega}\\
        D= \frac{\omega_{\bot}}{\omega}\\
        \phi =0
   \end{cases}$

# The spin motion is given by:

# $\begin{cases} \begin{align*} S_x &= -\frac{\omega_{//}}{\omega} \sin\omega s \\ S_y&= \frac{{\omega_{//}}^2}{{\omega}^2} \cos\omega s+\frac{{\omega_{\bot}}^2}{{\omega}^2}\\ S_z& = -\frac{\omega_{//}\omega_{\bot}}{{\omega}^2} \big(1-\cos\omega s\big) \end{align*}\end{cases}$


# Now consider the random initial condition($\tilde{S}_x,\tilde{S}_y,\tilde{S}_z$)
# At S = 0:
# $\begin{cases} \begin{align*} \tilde{S}_x &= C\sin\phi \\ \tilde{S}_y &= -\frac{\omega_{//}}{\omega } C\cos\phi +\frac{\omega_{\bot}}{\omega }D\\ \tilde{S}_z&=\frac{\omega_{\bot}}{\omega } C\cos\phi+\frac{\omega_{//}}{\omega }D \end{align*}\end{cases}$

# Use the second and the third equation to eliminate $\phi$  , can get $D=\frac{\omega_{\bot}\tilde{S}_y+\omega_{//}\tilde{S}_z}{\omega }$

# Cancel D in the 2nd or 3rd equation can get:   

# $C \cos\phi=\frac{\omega_{\bot}\tilde{S}_z-\omega_{//}\tilde{S}_y}{\omega}$


# $C = \frac{\omega_{\bot}\tilde{S}_z-\omega_{//}\tilde{S}_y}{\omega} \cos\phi+ \tilde{S}_x \sin\phi $


# Since we know $\tilde{S}_x = C\sin\phi$, then $\phi= \arctan\big(\frac{\omega \tilde{S}_x}{\omega_{\bot}\tilde{S}_z-\omega_{//}\tilde{S}_y}\big)$


# Summarizing the information above can get:

# $\begin{cases}
    D = \frac{\omega_{\bot} \tilde{S}_y + \omega_{//} \tilde{S}_z}{\omega}\\
    \phi = \arctan \big (\frac{\omega \tilde{S}_x }{-\omega_{//} \tilde{S}_y + \omega_{\bot} \tilde{S}_z}\big)\\
    C = \frac{-\omega_{//}\tilde{S}_y+\omega_{\bot}\tilde{S}_z}{\omega} \cos\phi+ \tilde{S}_x \sin\phi 
\end{cases}$

```python
E= 7007290000          # electron energy 7GeV
m= 510998.95           # rest mass energy of electron 511KeV
a= 0.001159652         # anomalous g factor for electron
c= 299792458           # speed of light
L= 5.902202            # length of both the B2E，BLA4lLE， BLA6RE, unit is metre
B= 0.22075143          # dipole strength of B2E in Tesla
ga= E/m                #lorentz boost fact
SLa = -4.843           #Solen strength B2EAL
SLb = -2.577           #Solen Srength B2EBL
SRa = -3.608           #Solen strength B2EAR
SRb = -3.942           #Solen Srength B2EBR
```

```python
pi/a
```

```python
ga*a/(2*pi)
```

```python
def SR(B,S):
    Wb = (1+a*ga)*B*c/E
    Ws = (1+a)*S*c/E
    W = sqrt(Wb**2+Ws**2)
    return Wb,Ws,W
```

```python
def S_G(B,S,si,s):       #to find spin at the exit of the dipole and the solenoid combined function magnet
    
    six = si[0]        # intial spin.x
    siy = si[1]        # intial spin.y
    siz = si[2]        # intial spin.z
    
    wb,ws,w = SR(B,S)
         
    D = (wb*siy+ws*siz)/w  # find D
    
    #print('the amount of rotation is: %s degrees'%(w*s*180/pi))
            
    cc = (-ws*siy+wb*siz)/w  
    
    cs = six
        
    sx =  cc*sin(w*s)+cs*cos(w*s)
    
    sy = -ws/w*(cc*cos(w*s)-cs*sin(w*s))+wb/w*D
    
    sz =  wb/w*(cc*cos(w*s)-cs*sin(w*s))+ws/w*D
    
    return sx,sy,sz
```

```python
s0= array([0, 1, 0])        #initial spin which is in the y direction
```

```python
sg=S_G(B,SLb,s0,L)  # when particle passing through the B2EBL spin rotator 
print('the spin at the exit of B2EBL rot',sg)
```

```python
sg=S_G(B,SLa,sg,L)  # when particle passing through the B2EAL spin rotator 
print('the spin at the exit of B2EAL rot',sg)
```

```python
s0= array([0, 1, 0]) 
wblb,wslb,wlb = SR(B,SLb)
slb = arange(0, 2*pi/wlb+0.1, 0.1)
Sglb=S_G(B,SLb,s0,slb)  # when particle passing through the B2EBL spin rotator

slbL = arange(0, L+0.1, 0.1)
SglbL=S_G(B,SLb,s0,slbL)  # when particle passing through the B2EBL spin rotator 
```

```python
s0 = array([SglbL[0][-1],SglbL[1][-1],SglbL[2][-1]])
wbla,wsla,wla = SR(B,SLa)
sla = arange(0, 2*pi/wla+0.1, 0.1)
Sgla=S_G(B,SLa, s0,sla)  # when particle passing through the B2EAL spin rotator 

slaL = arange(0, L+0.1, 0.1)
SglaL=S_G(B,SLa,s0,slaL)  # when particle passing through the B2EAL spin rotator 
```
